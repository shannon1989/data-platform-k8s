## ✅ 方案一（最推荐）：注册 spark_catalog 为 Iceberg（默认）

这是 最干净、最工业化 的做法。

```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .config(
        "spark.sql.extensions",
        "org.apache.iceberg.spark.extensions.IcebergSparkSessionExtensions",
    )
    .config(
        "spark.sql.catalog.spark_catalog",
        "org.apache.iceberg.spark.SparkSessionCatalog",
    )
    .config(
        "spark.sql.catalog.spark_catalog.type",
        "hadoop",
    )
    .config(
        "spark.sql.catalog.spark_catalog.warehouse",
        "s3a://datalake/iceberg",
    )
    .config("spark.hadoop.fs.s3a.endpoint", "http://minio.airflow:9000")
    .config("spark.hadoop.fs.s3a.access.key", "minioadmin")
    .config("spark.hadoop.fs.s3a.secret.key", "minioadmin")
    .config("spark.hadoop.fs.s3a.path.style.access", "true")
    .config("spark.hadoop.fs.s3a.connection.ssl.enabled", "false")
    .config("spark.hadoop.fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem")
    .getOrCreate()
)
```

```python
spark.sql("SHOW CATALOGS").show()
```

```python
spark.sql("CREATE TABLE spark_catalog.default.test_iceberg (id int) USING iceberg")
```

```python
spark.sql("CREATE TABLE spark_catalog.default.test_iceberg_2 (id int) USING iceberg")
```

```python
spark.sql("select * from spark_catalog.default.test_iceberg_2").show()
```

```python
spark.sql("""
CREATE TABLE spark_catalog.default.t2 (
  id INT,
  name STRING
) USING iceberg
""")

spark.sql("INSERT INTO spark_catalog.default.t2 VALUES (1, 'a')")
spark.sql("SELECT * FROM spark_catalog.default.t2").show()
```
