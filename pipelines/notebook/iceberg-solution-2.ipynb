## ✅ 方案二：注册一个「独立 Iceberg catalog」（更直观）

- 如果你希望在 SHOW CATALOGS 里看到 iceberg

```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("iceberg-minio-test")
    # Iceberg
    .config(
        "spark.sql.extensions",
        "org.apache.iceberg.spark.extensions.IcebergSparkSessionExtensions",
    )
    .config(
        "spark.sql.catalog.iceberg",
        "org.apache.iceberg.spark.SparkCatalog",
    )
    .config("spark.sql.catalog.iceberg.type", "hadoop")
    .config(
        "spark.sql.catalog.iceberg.warehouse",
        "s3a://datalake/iceberg",
    )
    # MinIO / S3A
    .config("spark.hadoop.fs.s3a.endpoint", "http://minio.airflow.svc.cluster.local:9000")
    .config("spark.hadoop.fs.s3a.access.key", "minioadmin")
    .config("spark.hadoop.fs.s3a.secret.key", "minioadmin")
    .config("spark.hadoop.fs.s3a.path.style.access", "true")
    .config("spark.hadoop.fs.s3a.connection.ssl.enabled", "false")
    .config("spark.hadoop.fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem")
    .getOrCreate()
)
```

```python
spark.sql("SHOW CATALOGS").show()
```

```python
spark.sql("""
CREATE TABLE iceberg.default.t4 (
  id INT,
  name STRING
) USING iceberg
""")
```

```python
spark.sql("INSERT INTO iceberg.default.t4 VALUES (4, 'xiaolong')")
```
