```python
import json
def pretty_print_block(block: dict, max_len: int = 120):
    def shrink(v):
        if isinstance(v, str) and len(v) > max_len:
            return v[:max_len] + "...<truncated>"
        return v

    cleaned = {k: shrink(v) for k, v in block.items()}
    print(json.dumps(cleaned, indent=2, ensure_ascii=False))
```
