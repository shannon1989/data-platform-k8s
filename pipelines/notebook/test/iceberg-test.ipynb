### 1️⃣ Create namespace（database/schema）

```python
spark.sql("""
CREATE TABLE silver.eth_blocks_t01 (
  id INT,
  name STRING
) USING iceberg
""")

spark.sql("INSERT INTO silver.eth_blocks_t01 VALUES (1, 't01')")
spark.sql("SELECT * FROM silver.eth_blocks_t01").show()
```
